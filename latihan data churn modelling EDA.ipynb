```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"C:\Users\SHELY MULIAWATI\Desktop\materi bootupai\data\Bank_churn_modelling.csv")
for col in ["CreditScore","Age","NumOfProducts","Balance"]:
    plt.figure(figsize=(10,4))
    plt.scatter(df.index,df[col])
    plt.title(col)
    plt.show()
```

```python
plt.scatter([df.Age],[df.Balance])
```

```python
plt.figure(figsize=(12,5))
sns.distplot(df['CreditScore'][df.Exited==0])
sns.distplot(df['CreditScore'][df.Exited==1])
plt.legend(['0','1'])
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.distplot(df['Age'][df.Exited==0])
sns.distplot(df['Age'][df.Exited==1])
plt.legend(['0','1'])
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.distplot(df['Balance'][df.Exited==0])
sns.distplot(df['Balance'][df.Exited==1])
plt.legend(['0','1'])
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.distplot(df['NumOfProducts'][df.Exited==0])
sns.distplot(df['NumOfProducts'][df.Exited==1])
plt.legend(['0','1'])
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.distplot(df['NumOfProducts'][df.Exited==0])
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.distplot(df['NumOfProducts'][df.Exited==1])
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.distplot(df['CreditScore'][df.Geography=='France'])
sns.distplot(df['CreditScore'][df.Geography=='Spain'])
plt.legend(['France','Spain'])
plt.show()
```

```python
# Geography v/s Exited
plt.figure(figsize=(6,3))
sns.countplot(df.Geography)
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.countplot(df.Geography[df.Exited==1])
plt.show()
```

```python
# Gender v/s Exited
plt.figure(figsize=(6,3))
sns.countplot(df.Gender)
plt.show()

plt.figure(figsize=(6,3))
sns.countplot(df.Gender[df.Exited==1])
plt.show()
```

```python
# HasCrCard v/s Exited
plt.figure(figsize=(6,3))
sns.countplot(df.HasCrCard)
plt.show()

plt.figure(figsize=(6,3))
sns.countplot(df.HasCrCard[df.Exited==1])
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x="Age",y="CreditScore",hue="Exited",data=df)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.pointplot(x="Geography",y="Age",hue="Exited",data=df)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.pointplot(x="NumOfProducts",y="Balance",hue="Exited",data=df)
plt.show()
```

```python
plt.figure(figsize=(5,7))
sns.scatterplot(x="NumOfProducts",y="Balance",hue="Exited",data=df)
plt.xticks([1,2,3,4])
plt.show()
```

```python
cor = df.corr()
cor
```
